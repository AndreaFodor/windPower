```python
weather_files = os.listdir('../data/windfarm_weather_data') 
```

```python

for farm_name in weather_files:
    farm_wdf = pd.read_csv(f"../data/windfarm_weather_data/{farm_name}")
    farm_wdf['Date/Time'] = pd.to_datetime(farm_wdf['time'])
    farm_wdf['Year'] = farm_wdf['Date/Time'].apply(lambda x: x.year)
    farm_counts_per_year = [0]*6
    for i, year in enumerate([2019,2020,2021,2022,2023,2024]):
        farm_counts_per_year[i] = len(farm_wdf[farm_wdf['Year'].astype(int) == year])
    if farm_counts_per_year == [8760, 8784, 8760, 8760, 8760, 8784]:
        print(f"{farm_name.replace(' wind farm hourly weather 2019-2024.csv','')}: Count matches")
    else:
        print(f"{farm_name.replace(' wind farm hourly weather 2019-2024.csv','')}: Count does not match")
```
